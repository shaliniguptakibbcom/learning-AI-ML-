```python
#Generate 10000 random normally distributed numbers and plot the density curve along with the median and mean
norm_data = pd.DataFrame(np.random.normal(size=100000))
norm_data.plot(kind="density",figsize=(10,6))
plt.vlines(norm_data.mean(), ymin=0, ymax=0.8, linewidth=5.0)
plt.vlines(norm_data.median(),
ymin=0,
ymax=0.8,
linewidth=5.0,
color="red")
```

```python
skewed_data = pd.DataFrame(np.random.exponential(size=100000))

skewed_data.plot(kind="density",figsize=(10,10),xlim=(-1,5))
plt.vlines(skewed_data.mean(), ymin=0, ymax=0.8, linewidth=5.0)
plt.vlines(skewed_data.median(),
ymin=0,
ymax=0.8,
linewidth=5.0,
color="red")
```
